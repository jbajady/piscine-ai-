```python
import numpy as np 
import pandas as pd
```

```python
data =pd.read_csv("data/Ecommerce_purchases.txt",
 sep=","
 ,low_memory=False)
print(data.shape)
data.head()
```

```python
avg_price = data['Purchase Price'].mean()
print(avg_price)
```

```python
max_price = data['Purchase Price'].max()
min_price = data['Purchase Price'].min()
print("max: ",max_price, "\nmin: ",min_price)
```

```python
pepol_chose_en=data["Language"].value_counts()["en"]
print(pepol_chose_en)
```

```python
cout_job_Lawyer=data["Job"].value_counts()["Lawyer"]

print(cout_job_Lawyer)
```

```python
am=data["AM or PM"].value_counts()["AM"]
pm=data["AM or PM"].value_counts()["PM"]

print("AM: ",am,"PM: ",pm)
```

```python
types_job=data['Job'].value_counts().head(5)
print(types_job)
```

```python
result = data[data['Lot'] == '90 WT']['Purchase Price']
print(result)
```

```python
email=data[data["Credit Card"] == 4926535242672853]["Email"]
print(email)
```

```python

count = data[(data["Language"] == "en") & (data["Purchase Price"] > 95)].shape[0]
print(f"Number of people: {count}")
```
