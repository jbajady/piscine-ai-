```python
import numpy as np
import pandas as pd 
```

```python
data =pd.read_csv("data/household_power_consumption.txt",
 sep=";"
 ,low_memory=False)
data
```

```python
data.drop(["Time", "Sub_metering_2", "Sub_metering_3"], axis=1, inplace=True)
data.head()
```

```python
data.set_index("Date",inplace=True)
data.index = pd.to_datetime(data.index, dayfirst=True)
print(data.index.dtype)
data.head()
```

```python
def update_types(df):
        for col in df.columns :
            df[col]=pd.to_numeric(df[col],errors="coerce")
        return df
        
data = update_types(data)
print(data.dtypes)
```

```python
data.describe()
```

```python
print(data.isna().sum())
data.dropna(inplace=True)
print(data.isna().sum())
```

```python
# data["Sub_metering_1"] = pd.to_numeric(data["Sub_metering_1"], errors="coerce")
data.loc[:, "Sub_metering_1"]=(data["Sub_metering_1"]+1)*0.06
data["Sub_metering_1"].head()
```
