```python
import plotly.express as px
import numpy as np 
import pandas as pd


returns = np.random.randn(50)
price = 100 + np.cumsum(returns)

dates = pd.date_range(start='2020-09-01', periods=50, freq='B')
df = pd.DataFrame(zip(dates, price),
                  columns=['Date','Company_A'])

fig = px.line(df, x='Date', y="Company_A",title="Company_1stock price")

fig.show()
df.head()
```

```python
import plotly.graph_objects as go

fig = go.Figure([go.Scatter(x=df['Date'], y=df['Company_A'])])
fig.update_layout(title="Company_1stock price", xaxis_title="Date", yaxis_title="Company_A")
fig.show()
```
