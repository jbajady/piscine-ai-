```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame({
        'name':['christopher','marion','maria','mia','clement','randy','remi'],
        'age':[70,30,22,19,45,33,20],
        'gender':['M','F','F','F','M','M','M'],
        'state':['california','dc','california','dc','california','new york','porto'],
        'num_children':[2,0,0,3,8,1,4],
        'num_pets':[5,1,0,5,2,2,3]
        })
print(df)
```

```python
df.plot(
    x='name',                        # x-axis: names
    y=['num_children', 'num_pets'],  # y-axis: two columns
    kind='bar',                      # bar plot
    figsize=(10, 6),                 # figure size
    title='Number of children and pets per person' # title
)

plt.xlabel('name')        # x-axis label
plt.ylabel('value')       # y-axis label
plt.xticks(rotation=45)   # rotate names for readability
plt.legend(['num_children', 'num_pets'])  # legend
plt.tight_layout()        # avoid clipping
plt.show()
```
