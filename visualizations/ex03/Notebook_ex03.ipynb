```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
        'x-axis':[1,4,5,6,7],
        'y-axis':[2,6,3,6,3]
        })
df.head()

plt.plot(
    df["x-axis"], df["y-axis"],
    linestyle=':',   
    color='red',  
    linewidth=3,  
    marker='o',    
    markerfacecolor='blue',  
    markersize=12,
    # title="ddddd"
    
)
plt.xlim(1,8)
plt.ylim(1,8)
plt.title("Display marker")
plt.xlabel("x -axis")
plt.ylabel("y -axis")
plt.show()
```
