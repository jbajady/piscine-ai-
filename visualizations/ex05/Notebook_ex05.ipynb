```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3)

fig.subplots_adjust(hspace=0.5, wspace=0.5)

axes = axes.flatten()

for i in range(6):
    axes[i].text(
        0.5, 0.5, f"(2, 3, {i+1})",
        ha='center', va='center', fontsize=12
    )
    axes[i].set_title(f"Title {i+1}")

plt.show()
```
