```python
import numpy as np

data = np.genfromtxt(
    "data/winequality-red.csv",
    delimiter=';',
    dtype=np.float32
)

print(data.nbytes)

# Display rows 2, 7, 12
rows = data[[1, 6, 11]]

rows = rows[~np.isnan(rows)].reshape(3, -1)
print(rows)

# alcohol > 20%
alcohol = data[:, 10]  
result = np.any(alcohol > 20)
print(result)

#  average alcohol
avg_alcohol = np.nanmean(alcohol)
print(avg_alcohol)

# pH statistics
ph = data[:, 8]

print("min:", np.nanmin(ph))
print("max:", np.nanmax(ph))
print("25%:", np.nanpercentile(ph, 25))
print("50%:", np.nanpercentile(ph, 50))
print("75%:", np.nanpercentile(ph, 75))
print("mean:", np.nanmean(ph))

# average quality of lowest 20% sulphates
sulphates = data[:, 9]
quality = data[:, 11]

threshold = np.nanpercentile(sulphates, 20)
mask = sulphates <= threshold

avg_quality = np.nanmean(quality[mask])
print(avg_quality)

# best
max_q = np.nanmax(quality)
best_mask = quality == max_q
best_mean = np.nanmean(data[best_mask], axis=0)

# worst
min_q = np.nanmin(quality)
worst_mask = quality == min_q
worst_mean = np.nanmean(data[worst_mask], axis=0)

print("Best mean:", best_mean)
print("Worst mean:", worst_mean)

```
